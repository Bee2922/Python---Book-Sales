```python
# To create a histogram to see the distribution of the publishing year
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
# To show the number of books in each genre, we are going to first of all view the count of the genre column before plotting the bar chart
```

```python
df["genre"].value_counts()
```

```python
df["genre"].value_counts().plot(kind = "bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in each Genre")
plt.show()
```

```python
## Group the books by author and their ratings using boxplot
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x = "genre", y = "Book_ratings_count", data = df)
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Box Plot of Book Ratings Count for Each Genre") 
plt.show()
```

```python
# Create a scattered plots of sale price and units sold
```

```python
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter Plot of Sale Price vs Units Sold")
plt.show()
```

```python
# The Scatter plot above shows that more units of books were sold when the prices were low, which is the exact opposite in the case of high prices
```

```python
#Create a pie chart of the language distribution. First of all, it is helpful to count the available languages, followed by plotting the pie chart. You have to define what language counts is, which is represented by the code below
```

```python
df["language_code"].value_counts()
```

```python
language_counts = df["language_code"].value_counts()
plt.pie(language_counts, labels = language_counts.index, startangle = 90)
plt.title("Language Distribution of Books")
plt.show()
```

```python
From the pie chart above, it clearly shows that the books written in general english sold more than books written in other languages
```

```python
## Group the books by publisher and calculate the total revenue for each publisher
```

```python
df.groupby("Publisher ")["publisher revenue"].sum(). sort_values(ascending = False)
```

```python
# Create a barchart of total gross sales for each author. First thing to do is to define what total gross sales by author is as shown below
```

```python
Total_gross_sales_by_authors = df.groupby("Author")["gross sales"].sum(). sort_values(ascending = False).head(15).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales per Author")
plt.show()
```

```python
Total_gross_sales_by_authors
```

```python
## One could clearly see the gross sales of the top 10 authors from the bar chart above, arranged in descending order
```

```python
# Create a boxplot of unit sold by each author ratings
```

```python
sns.boxplot(x = "Author_Rating", y = "units sold", data = df)
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.title("Boxplot of Unit sold vs author ratings")
plt.show()
```

```python
## Group data by publishing year and plot a line chart of total units sold over the years
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind = "line", marker = "o")
plt.xlabel("Publishing Year")
plt.ylabel("Unites Sold")
plt.title(" Line Chart of Publishing Year vs Units Sold")
plt.show()
```

```python
## The line chart above shows that more units of books were sold in Year 2000, with the 800,000 units shown as the highest unit of books sold
```
